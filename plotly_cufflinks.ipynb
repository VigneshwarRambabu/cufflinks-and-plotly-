```python
#Choropleth on real data
```

```python
df = pd.read_csv('2011_US_AGRI_Exports')
```

```python
df.head()
```

```python
df.info()
```

```python
 df.head()
```

```python
data = dict(type='choropleth',
           locations=df['code'],
           locationmode='USA-states',
           text=df['state'],
           colorscale='Reds',
           marker = dict(line=dict(color='rgb(255,255,255)',width=1)),
           z=df['total exports'],
           colorbar = {'title':"Total exports of US regions"})
```

```python
 layout = dict( title ='2011 US AGRICULTURE EXPORTS BY STATE',
               geo=dict(scope= 'usa', showlakes=True,lakecolor='rgb(85,173,240)'))
```

```python
chloromap1 = go.Figure(data= [data],layout= layout)
```

```python
iplot(chloromap1)
```

```python
df = pd.read_csv('2014_World_GDP')
```

```python
df.head()
```

```python
data=dict(type='choropleth',
         locations=df['CODE'],
          text=df['COUNTRY'],colorscale = 'Viridis',
         z=df['GDP (BILLIONS)'],
          colorbar={'title':'GDP in BILLIONS USD'}
         )
```

```python
layout=dict(title='2014_World_GDP', geo=dict(showlakes=False,projection={'type':'orthographic'}))
```

```python
go.Figure(data=[data],layout= layout)
```
